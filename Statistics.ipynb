```python
import pandas as pd
import numpy as np

import seaborn as sns

from matplotlib import pyplot as plt
```

```python
sns.set_theme(style='darkgrid')
```

```python
X = pd.read_csv('googleplaystore.csv')
```

```python
X.head()
```

```python
X.Type.value_counts()
```

```python
X = X[X.Type != '0'].reset_index(drop=True)
# Убираем строку с некорректными данными
```

```python
X.Type.value_counts()
```

```python
#X.Category.value_counts().plot(kind='bar', figsize=(16, 6))
X.Category.value_counts().plot.bar(figsize=(16, 6))
plt.title('Categories')
# Три самых популярных категории: FAMILY, GAME, TOOLS
```

```python
X.dtypes
```

```python
X.Rating.plot.box(figsize=(14, 6))
plt.title('Rating boxplot')
# Наибольшее количество оценок (Rating) располагаются в диапазоне от 4 до 4.5
```

```python
X.Price = X.Price.str.extract('\$?(.*)')[0].astype(float)
# отсекаем знвк $
```

```python
X.Price[X.Price != 0].plot.hist(bins=50, figsize=(14, 5))
plt.xlabel('Price')
plt.ylabel('Freq')
plt.title('Price histogram')
# Большое количество приложений бесплатных, из платных приложений основная часть стоит в диапазоне от $0.01, до $50. Часть приложений очень дорогих ($350-$400).
```

```python
X[X.Price > 350]
```

```python
X.Price[X.Price != 0].plot.box(figsize=(14, 5))
# Большое количество приложений бесплатных, из платных приложений основная часть стоит в диапазоне от $0.01, до $50. Часть приложений очень дорогих ($350-$400).
```

```python
sns.kdeplot(data=X, x='Rating')
plt.title('Rating KDE - plot')
# Максимальное количество приложений с оценкой ~4.3
```

```python
X['Content Rating'].value_counts()
```

```python
sns.kdeplot(data=X, x='Rating', hue='Content Rating')
# Вывод рейтинга по категориям. Вне зависимости от категории приложения максимальное количество оценок с рейтингом ~4.3.
```

```python
sns.boxplot(data=X, y='Rating', x='Content Rating')
# Вывод рейтинга по категориям. Вне зависимости от категории приложения максимальное количество оценок с рейтингом ~4.3.
```

```python
fig, axis = plt.subplots(1, 3, figsize=(18, 6),  sharey=True)

step = 30
P = X[X.Price > 0]
for i in range(0, 3):
  price_from = i*step
  price_to = price_from + step
  sns.kdeplot(data=P[(P.Price >= price_from) & (P.Price < price_to)], x='Price', hue='Content Rating', ax=axis[i])

plt.tight_layout()
# В диапазоне цен от $20 до $100 предобладают приложения Content Rating = Everyone, остальные приложения есть только в ценовом диапазоне до $40, при рассмотрении диапазона от $0 до $100.
```

```python
X.groupby('Content Rating').Price.mean().sort_values(ascending=False).plot.bar()
plt.title('AVG Price / Content Rating')
plt.ylabel('AVG Price')
# Средняя цена категории Evereone выше остальных категорий
```

```python
v = X[X.Category.isin(['FAMILY', 'TOOLS', 'GAME'])].pivot_table(index='Content Rating', columns='Category', values='Price', aggfunc='mean').round(3)
v
# наиболее дорогие приложения в категориях FAMILY, TOOLS, GAME
```

```python
sns.heatmap(v, annot=True)
# Визуализация наиболее дорогих приложений в вышеуказанных категориях. Для FAMILY = Everyone, для GAME = Teen, для TOOLS = Mature 17+
```

```python
v.plot.bar()
# Визуализация наиболее дорогих приложений в вышеуказанных категориях. Для FAMILY = Everyone, для GAME = Teen, для TOOLS = Mature 17+
```

```python
X['date'] = pd.to_datetime(X['Last Updated'])
```

```python
v = X[X.Price > 0].groupby('date').Price.mean()
```

```python
v.cumsum().plot(figsize=(18, 8))
plt.title('Date/Price')
# Рост средних цен по годам выхода обновлений приложений. Соответственно, если цена низкая, значит, скорее всего приложение давно не обновлялось
```
